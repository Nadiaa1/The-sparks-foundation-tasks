```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Number of attacks according to Regions

```python
plt.figure(figsize=(20,14))
sns.barplot(x=data['region_txt'].value_counts().index,y=data['region_txt'].value_counts())
plt.xticks(rotation=90,fontsize=14)
plt.title('Number of attacks according to Regions')
plt.xlabel('Region')
plt.ylabel('Number of attaks')
plt.show()
```

# Top 30 countries got attaked

```python
plt.figure(figsize=(20,14))
sns.barplot(x=data['country_txt'].value_counts()[:30].index,y=data['country_txt'].value_counts()[:30])
plt.xticks(rotation=90,fontsize=14)
plt.title('Top 30 countries got attaked')
plt.xlabel('Country')
plt.ylabel('Number of attaks')
plt.show()
```

# Attacks types distribution

```python
plt.figure(figsize=(20,14))
sns.barplot(x=data['attacktype1_txt'].value_counts().index,y=data['attacktype1_txt'].value_counts())
plt.xticks(rotation=90,fontsize=14)
plt.title('attacks types distribution')
plt.xlabel('attack type')
plt.ylabel('number of attacks')
plt.show()
```

# Number of successful attacks according to Regions

```python
succ['number of successful attacks']=data.groupby('region_txt')['success'].transform('sum')
succ
```

# Number of attacks according to years

```python
plt.figure(figsize=(20,6))
sns.barplot(x=data['iyear'].value_counts()[:30].index,y=data['iyear'].value_counts()[:30])
plt.xticks(rotation=90,fontsize=14)
plt.title('Number of attacks according to years')
plt.xlabel('Year')
plt.ylabel('Number of attaks')
plt.show()
```
